```python
%matplotlib inline
import numpy as np
import pandas as pd
import scipy.spatial
import matplotlib.pyplot as plt
from astropy.io import fits
```

## Problem 1

```python
m81 = fits.getdata('m81.fits')
m87 = fits.getdata('m87.fits')
```

```python
def brightest_pixel(image):
    return np.unravel_index(image.argmax(), image.shape)
```

```python
def weighted_pixel(image):
    imean = np.dot(np.arange(image.shape[0]), image).sum()/image.sum()
    jmean = np.dot(image, np.arange(image.shape[1])).sum()/image.sum()
    return (int(round(imean)), int(round(jmean)))
```

### M81

```python
brightest_m81 = brightest_pixel(m81)
weighted_m81 = weighted_pixel(m81)
print brightest_m81, weighted_m81
```

```python
plt.imshow(m81)
plt.xlim(m81.shape[0]/2 - 15, m81.shape[0]/2 + 15)
plt.ylim(m81.shape[1]/2 - 15, m81.shape[1]/2 + 15)
plt.scatter(*zip(*[brightest_m81, weighted_m81]), marker='x')
plt.title('M81 Weighted Center vs. Brightest Pixel')
```

In this case, the brightest pixel appears to be the best indicator of the center of the galaxy. The weighted center is slightly off, likely because it has been swayed down and left by the bright spiral arms of the galaxy, which could be biased in that direction because of a tilt relative to our perspective.

### M87

```python
brightest_m87 = brightest_pixel(m87)
weighted_m87 = weighted_pixel(m87)
print brightest_m87, weighted_m87
```

```python
plt.imshow(m87)
plt.xlim(m87.shape[0]/2 - 15, m87.shape[0]/2 + 15)
plt.ylim(m87.shape[1]/2 - 15, m87.shape[1]/2 + 15)
plt.scatter(*zip(*[brightest_m87, weighted_m87]), marker='x')
plt.title('M87 Weighted Center vs. Brightest Pixel')
```

In this case, both methods are similar and only differ diagonally by 1 pixel. It is  difficult to judge based on this image alone where the center actually is, but the weighted center does seem to be slightly too far left due to the lump of brightness near the upper left of center.

## Problem 2

```python
def brightness_profile(image, center, nbins=100):
    # Calculate each index's distance from the center point.
    indices = np.array([[i, j] for j in range(image.shape[1]) for i in range(image.shape[0])])
    distances = scipy.spatial.distance.cdist(indices, np.atleast_2d(center)).ravel()
    
    # Calculate brightness profile, using image data to weight pixel values.
    brightness, bin_edges = np.histogram(distances, weights=image.flatten(), bins=nbins)
    base, bin_edges =  np.histogram(distances, bins=nbins)
    profile = brightness/base
    
    # Generate the exponential profile prediction.
    radius = np.linspace(0, 1, nbins)
    S_0 = profile[0]
    S = S_0*np.exp(-radius/0.04)
    
    return pd.DataFrame({'brightness': profile, 'exponential': S, 'radius': radius})
```

### M81

```python
m81_brightness = brightness_profile(m81, brightest_m81)
m81_brightnessplot = m81_brightness.plot(x='radius', y=['brightness', 'exponential'], title='M81 Brightness Profile')
lines, labels = m81_brightnessplot.get_legend_handles_labels()
m81_brightnessplot.legend(lines, ('Observed', 'Predicted'))
m81_brightnessplot.set_xlabel('Radius')
m81_brightnessplot.set_ylabel('Brightness')
```

### M87

```python
m87_brightness = brightness_profile(m87, brightest_m87)
m87_brightnessplot = m87_brightness.plot(x='radius', y=['brightness', 'exponential'], title='M87 Brightness Profile')
lines, labels = m87_brightnessplot.get_legend_handles_labels()
m87_brightnessplot.legend(lines, ('Observed', 'Predicted'))
m87_brightnessplot.set_xlabel('Radius')
m87_brightnessplot.set_ylabel('Brightness')
```

## Problem 3

```python
def cropped_stats(image, size=20):
    image = image[:size, :size].flatten()
    
    gaussian = np.random.normal(loc=np.mean(image), scale=np.std(image), size=len(image))
    poisson = np.random.poisson(size=len(image))
    
    return pd.DataFrame({'Image': image, 'Gaussian': gaussian, 'Poisson': poisson})
```

### M81

```python
m81_cropped = cropped_stats(m81)
m81_cropped.hist(layout=(1,3), figsize=(15,5))
m81_cropped['Image'].describe()
```

The Gaussian distribution is clearly the better fit.

### M87

```python
m87_cropped = cropped_stats(m87)
m87_cropped.hist(layout=(1,3), figsize=(15,5))
m87_cropped['Image'].describe()
```

Again, the Gaussian distribution is the better fit.
