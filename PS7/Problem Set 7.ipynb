```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.integrate import quad
from scipy.special import gammaincc 
```

## Problem 1

```python
lightcurve = pd.read_csv('KOI97.01_1.out', sep='\s*', names=['time', 'flux', 'flux_err'])
eclipse = lightcurve[(lightcurve['time'] > 124) & (lightcurve['time'] < 125)]
```

```python
eclipse['flux'].std(), eclipse['flux'].mean()
```

```python
def clip_sigma(data, deviations=2):
    # Return only data within the specified number of standard deviations.
    sigma = abs(data - data.mean())/data.std()
    return data[sigma < deviations]
```

```python
# Clip data to the mean unobscured stellar flux.
clipped_flux = clip_sigma(eclipse['flux'])
for _ in range(4):
    clipped_flux = clip_sigma(clipped_flux)
    
# Normalize to the mean unobscured stellar flux.
eclipse[['flux','flux_err']] = eclipse[['flux','flux_err']]/clipped_flux.mean()
```

```python
# Derive z from t
def z(t_0, tau):
    return lambda t: abs(t - t_0)/tau
```

```python
# Flux integration formulas with limb darkening.
def delta(r, p, z):
    z = abs(z)
    if (r >= z+p) or (r <= z-p):
        return 0
    elif r+z <= p:
        return 1
    else:
        return (1/np.pi)*(np.arccos((z**2 - p**2 + r**2)/(2*z*r)))

def I(r):
    return 1.0 - (1.0 - ((1.0-r**2)**(1/2.0))**(1/2.0))

def integrate_flux(p, a=0.0, b=1.0):
    return lambda z: quad(lambda r: 2.0*r*I(r)*(1.0-delta(r, p, z)), a, b)[0]/quad(lambda r: 2*r*I(r), a, b)[0]
```

```python
def generate_fit(p, t_0, tau):
    # Generate a fit given model parameters.
    eclipsefit = eclipse.copy()
    eclipsefit['z'] = eclipsefit['time'].apply(z(t_0, tau))
    eclipsefit['flux_fit'] = eclipsefit['z'].apply(integrate_flux(p))
    
    return eclipsefit

def goodness_of_fit(data, original='flux', fit='flux_fit', error='flux_err', params=3):
    # Calculate chi squared and p-value.
    chi2 = (((data[original] - data[fit])/data[error])**2).sum()
    pvalue = gammaincc((len(data) - params)/2.0, chi2/2.0)
    
    return {'chi2': chi2, 'p': pvalue}
```

```python
guess_fit = generate_fit(p=0.078, t_0=124.51, tau=0.1)
guess_stats = goodness_of_fit(guess_fit)
guess_stats['chi2']
```

## Problem 2

```python
guessplot = guess_fit.plot(x='time', y=['flux', 'flux_fit'], figsize=(15,10))
guessplot.set_title('Eclipse Guess Fit', fontsize=18)
guessplot.set_xlabel('Time (BKJD)', fontsize=14)
guessplot.set_ylabel('Normalized Flux Ratio', fontsize=14)
```

Depending on how accurate we're expecting our fit to be, this is not a good fit. The curve misses the normalized non-eclipse points by a few percent, and the lines of the eclipse are just slightly off, though the general shape of the fit does look correct.

```python
guess_stats['p']
```

The probability of getting this fit by chance is extremely low, which suggests this is a bad fit. This is also hinted at by the  high $\chi^2$ value.

```python
tau_values = np.linspace(0.08, 0.13, 6)
tau_variations = [{'params': {'\\tau': tau}, 'fit': generate_fit(0.078, 124.51, tau)} for tau in tau_values]
```

```python
# Plot results with parameters and goodness of fit.
taufig, tauaxes = plt.subplots(2,3,figsize=(15,10))
for i, result in enumerate(tau_variations):
    title = ' '.join('${0}$'.format('='.join((param, str(value)))) for param, value in result['params'].iteritems())
    stats = goodness_of_fit(result['fit'])
    title += ' $p='+str(stats['p'])+'$ $\chi^2='+str(stats['chi2'])+'$'
    plot = result['fit'].plot(x='time', y=['flux', 'flux_fit'], title=title, ax=tauaxes[i/3,i%3])
```

```python
chi2_values = [goodness_of_fit(result['fit'])['chi2'] for result in tau_variations]
```

```python
plt.plot(tau_values, chi2_values)
plt.plot(tau_values, [492.58583943297691+1.0]*len(tau_values))
plt.xlim([0.10-0.000008, 0.10+0.0000068])
plt.ylim([492.5,494])
```

The best fit is given by $\tau= 0.1$ $(\chi^2=492.586)$ and $4.345\cdot 10^{-76}$. The one-sigma uncertainty of $\tau$ is $0.099992-0.1000068$, or approximately $0.1 \pm 0.0000074$.

## Problem 4

```python
p_values = np.linspace(0.0765, 0.0785, 21)
t_0_values = np.linspace(124.50, 124.55, 21)
tau_values = np.linspace(0.09, 0.11, 17)
```

```python
results = []
for p in p_values:
    for t_0 in t_0_values:
        for tau in tau_values:
            result = {'params': {'p':p,'t_0':t_0,'\\tau': tau}, 'fit': generate_fit(p, t_0, tau)}
            result['stats'] = goodness_of_fit(result['fit'])
            results.append(result)
```

```python
# Find best fit based on lowest chi sqaured value.
chi2_min = np.inf
for result in results:
    if result['stats']['chi2'] < chi2_min:
        chi2_min = result['stats']['chi2']
        best_fit = result
```

```python
# Plot best fit and details.
title = 'Eclipse Best Fit, '
title += ' '.join('${0}$'.format('='.join((param, str(value)))) for param, value in result['params'].iteritems())
title += '  $p-value='+str(best_fit['stats']['p'])+'$ $\chi^2='+str(best_fit['stats']['chi2'])+'$'
bestfitplot = best_fit['fit'].plot(x='time', y=['flux', 'flux_fit'], figsize=(15,10))
bestfitplot.set_title(title, fontsize=18)
bestfitplot.set_xlabel('Time (BKJD)', fontsize=14)
bestfitplot.set_ylabel('Normalized Flux Ratio', fontsize=14)
```
