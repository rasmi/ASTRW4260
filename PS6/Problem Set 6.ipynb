```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Useful constants.
G = 6.67e-11
AU = 1.496e11
seconds_year = 3.154e7

def planets(*args):
    # Returns planet objects with mass and
    # position, velocity, and acceleration vectors.
    
    def initialize_position(body):
        # Initialize the position vector of a body given
        # semimajor axis in AU and eccentrity of orbit.
        body['r'][1] = body['d']*AU*np.sqrt(1.0 - body['e']**2)

    def initialize_velocity(body):
        # Initialize tangential velocity to an eccentric orbit about the sun
        # at pericenter. We're ignoring contributions from other bodies.
        body['v'][0] = np.sqrt((G*(planets['sun']['m']+body['m'])*(1.0+body['e']))/(np.linalg.norm(body['r'])*(1.0-body['e'])))
    
    planets = {
        'sun': {
            'm': 1.988435e30,
            'r': np.zeros(3),
            'v': np.zeros(3),
            'a': np.zeros(3)
        },
        'earth': {
            'e': 0.0,
            'd': 1.0,
            'm': 5.9721986e24,
            'r': np.zeros(3),
            'v': np.zeros(3),
            'a': np.zeros(3)
        },
        'jupiter': {
            'e': 0.6,
            'd': 4.0,
            'm': 1.89813e27,
            'r': np.zeros(3),
            'v': np.zeros(3),
            'a': np.zeros(3)
        }
    }
    
    initialize_position(planets['earth'])
    initialize_position(planets['jupiter'])
    initialize_velocity(planets['earth'])
    initialize_velocity(planets['jupiter'])
    
    return [planets[planet] for planet in args]
```

## Problem 1

```python
def simulate_nbody(bodies, years, dt):
    # Numerical n-body solver using leapfrog method given 
    # bodies, a time limit in years, and dt timestep in seconds.
    
    # Keep track of all body coordinates for plotting.
    x = np.array([])
    y = np.array([])
    
    # Convert time limit to seconds.
    timelimit = years*seconds_year
    t = 0.0
    
    # Monitor kinetic and potential energy.
    def pe(body):
        # Gravitational potential energy of a body.
        pe = 0.0
        for b in bodies:
            if b is not body:
                r = (b['r']-body['r'])
                pe += (G*body['m']*b['m'])/(np.linalg.norm(r))
        return pe
    
    ke_initial = sum([0.5*body['m']*np.linalg.norm(body['v'])**2 for body in bodies])
    pe_initial = sum([pe(body) for body in bodies])
    energy_initial = ke_initial + pe_initial
    
    def a(body):
        # Acceleration on body by all other bodies.
        a = np.zeros(body['a'].shape)
        for b in bodies:
            if b is not body:
                r = (b['r']-body['r'])
                a += (G*b['m']*r)/(np.linalg.norm(r)**3)
        return a
    
    # Initialize position-based accelerations.
    for body in bodies:
        body['a'] = a(body)
        body['v'] += 0.5*body['a']*dt
    
    while t < timelimit:
        # Leapfrog integrate one timestep.
        for body in bodies:
            body['r'] += body['v']*dt
            
        for body in bodies:
            body['a'] = a(body)
            body['v'] += body['a']*dt
        t += dt
        
        x = np.append(x,[body['r'][0] for body in bodies])
        y = np.append(y,[body['r'][1] for body in bodies])
    
    # Calculate final energy. 
    ke_final = sum([0.5*body['m']*np.linalg.norm(body['v'])**2 for body in bodies])
    pe_final = sum([pe(body) for body in bodies])
    energy_final = ke_final + pe_final
    print "Total energy difference: " + str(abs((energy_initial-energy_final)/energy_initial)*100) + '%'
    
    plt.figure(figsize=(10,10))
    plt.scatter(x, y, marker='.')
    
    return bodies
```

```python
sun_2body, earth_2body = simulate_nbody(planets('sun', 'earth'), 5, 10**5)
```

A timestep of $10^5$ seconds (about one day) retains accuracy -- the orbit of the earth has approximately the same magnitude after 5 orbits and maintains a nice circle. Total energy decreases by less than one hundredth of a percent (energy is conserved), so we can have some confidence in this choice of timestep.
