```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Problem 1

```python
# Useful constants
G = 6.67384*10**-11
r_earth = 6367444.7
m_earth = 5.9721986*10**24

material_densities = {
    'Fe': {'rho_0': 8300.0, 'c': 0.00349, 'n': 0.528},
    'MgSiO3': {'rho_0': 4100.0, 'c': 0.00161, 'n': 0.541},
    'H20': {'rho_0': 1460.0, 'c': 0.00311, 'n': 0.513}
}
```

```python
# Return a density function given fit parameters
def rhofit(rho_0, c, n):
    # Generates density at a certain pressure.
    return lambda P: rho_0 + c*P**n
```

```python
def RungeKutta(P, m, r, dr, rho):
    
    def dPdr(P, m, r):
        return (-G*rho(P)*m)/r**2
    
    def dmdr(P, r):
        return 4.0*np.pi*rho(P)*r**2
    
    k1 = dPdr(P, m, r)*dr
    l1 = dmdr(P, r)*dr
    k2 = dPdr(P + k1/2.0, m + l1/2.0, r + dr/2.0)*dr
    l2 = dmdr(P + k1/2.0, r + dr/2.0)*dr
    k3 = dPdr(P + k2/2.0, m + l2/2.0, r + dr/2.0)*dr
    l3 = dmdr(P + k2/2.0, r + dr/2.0)*dr
    k4 = dPdr(P + k3, m + l3, r + dr)*dr
    l4 = dmdr(P + k3, r + dr)*dr
    
    dP = (1/6.0)*(k1 + 2.0*k2 + 2.0*k3 + k4)
    dm = (1/6.0)*(l1 + 2.0*l2 + 2.0*l3 + l4)
    
    return dP, dm
```

```python
def simulatePlanet(P, material):
    
    # Get a density function using fit parameters.
    rho = rhofit(**material_densities[material])
    
    m = 0
    dr = r_earth/10**4
    r = dr

    planet = pd.DataFrame({'P': [P], 'r': [r], 'm': [m]})
    
    # Choose a tolerance value for P to cut off iteration as we approach P = 0.
    tolerance = 10**0

    while P > tolerance:
        try:
            dP, dm = RungeKutta(P, m, r, dr, rho)
            if P + dP > 0:
                # dP is not too large, so we can continue to iterate towards P = 0.
                P += dP
                m += dm
                r += dr
                planet = planet.append({'P': P, 'r': r, 'm': m}, ignore_index=True)
            else:
                raise ValueError('dr is too large, decreasing...')
        except ValueError:
            # Step is too large, decrease dr to approach P = 0 more accurately.
            dr *= 0.001
    
    planet['density'] = planet['P'].apply(rho)
    planet['m_earth'] = planet['m']/m_earth
    planet['r_earth'] = planet['r']/r_earth
    
    return planet
```

**Note:** Pandas's `DataFrame.append()` operation is expensive during iteration because a new object must be created every time a new row is appended. I used it for the conciseness and convenience of Pandas here, but in real-world use, simple NumPy arrays for each attribute would be more efficient during iteration (and they can easily be included in a DataFrame afterwards).

```python
planet = simulatePlanet(P=10**12, material='Fe')
planet.tail(1)
```

```python
densityplot = planet.plot(x='r_earth', y='density', title='Iron Planet Density vs. Radius', legend=False)
densityplot.set_xlabel('Radius $(R_\oplus)$')
densityplot.set_ylabel('Density $(kg \cdot m^{-3})$')
```

```python
simulateFe = [simulatePlanet(10**P,'Fe') for P in range(10,16)]
planetsFe = pd.DataFrame({
    'mass': [planet['m_earth'].iloc[-1] for planet in simulateFe],
    'radius': [planet['r_earth'].iloc[-1] for planet in simulateFe]
})
```

## Problem 3

```python
planets = {}

for planettype in material_densities.keys():
    simulated = [simulatePlanet(10**P, planettype) for P in range(10,16)]
    planets[planettype] = pd.DataFrame({
        'mass': [planet['m_earth'].iloc[-1] for planet in simulated],
        'radius': [planet['r_earth'].iloc[-1] for planet in simulated]
    })
```

```python
terrestrialplanets = {
    'Mercury': pd.DataFrame({'mass':[0.0553], 'radius':[0.383]}),
    'Venus': pd.DataFrame({'mass':[0.815], 'radius':[0.949]}),
    'Earth': pd.DataFrame({'mass':[1.0], 'radius':[1.0]}),
    'Mars': pd.DataFrame({'mass':[0.1075], 'radius':[0.5325]})
}

exoplanets = {
    'HD 97658 b': pd.DataFrame({'mass':[7.87], 'radius':[2.34]}),
    'Kepler-18 b': pd.DataFrame({'mass':[6.90], 'radius':[2.00]}),
    'Kepler-20 b': pd.DataFrame({'mass':[8.47], 'radius':[1.91]}),
    'Kepler-37 c': pd.DataFrame({'mass':[3.35], 'radius':[0.75]}),
    'Kepler-100 c': pd.DataFrame({'mass':[0.85], 'radius':[2.20]}),
    'Kepler-109 b': pd.DataFrame({'mass':[1.30], 'radius':[2.37]}),
    'Kepler-113 b': pd.DataFrame({'mass':[7.10], 'radius':[1.82]}),
    'Kepler-406 b': pd.DataFrame({'mass':[4.71], 'radius':[1.43]}),
    'KOI-94b': pd.DataFrame({'mass':[10.5], 'radius':[1.71]})
}

planets.update(terrestrialplanets)
planets.update(exoplanets)
```

```python
planetsplot = plt.figure(figsize=(18,10)).add_subplot(111)

for planet in planets.values():
    style = 'o' if len(planet) is 1 else ''
    planet.plot(x='mass', y='radius', ax=planetsplot, logx=True, style=style)

plt.xlabel('Mass $(M_\oplus)$', fontsize=18)
plt.ylabel('Radius $(R_\oplus)$', fontsize=18)
plt.title('Planet Mass-Radius Relations', fontsize=24)
plt.legend(planetsplot.get_lines(), planets.keys(), numpoints=1, loc='best')
```
